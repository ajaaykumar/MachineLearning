```python
import json
import csv
```

```python
def extend(dct):
    dct_key = list(dct.keys())
    for key in dct_key:
        if type(dct[key])==dict:
            key_lst = dct[key].keys()
            key_dct = {key+"."+x: dct[key][x] for x in key_lst}
            dct.update(key_dct)
            del dct[key]
    return dct
```

```python
with open("ipdataframe.json", "r") as read_file:
    data = json.load(read_file)
data = [extend(x) for x in data]
for d in data:
    d.update((k, '') for k, v in d.items() if v == [])
```

```python
def extend_keys(dct):
    dct["abuse_ch_category"] = "|".join([x.get("title",'') for x in dct["abuse_ch_category"]])
    dct["av_hits_list"]      = "|".join([x.get("name",'')+"^"+x.get("result",'') for x in dct["av_hits_list"]])
    dct["ti_data.feed"]      = "|".join([x.get("desc",'')+"^"+x.get("domain",'') for x in dct["ti_data.feed"]])
    
    dct["signatures"]        = "|".join([x.get("signature",'')+"^"+x.get("category")+"^"+x.get("severity") for x in dct["signatures"]])
    dct["signatures"]        = dct["signatures"].replace("&","&1").replace(",","&2")
    
    dct["dnsbl_blockedlist"] = "|".join(dct.get("dnsbl_blockedlist",''))
    dct["dnsbl_blockedlist"] = dct["dnsbl_blockedlist"].replace("&","&1").replace(",","&2") 
    
    dct["sensor_location"]   = dct["sensor_location"].replace("&","&1").replace(",","&2")
    
    dct["ti_tags"]           = "|".join(dct.get("ti_tags",''))
    dct["ti_tags"]           = dct["ti_tags"].replace("&","&1").replace(",","&2")
    
    dct["ti_data.passive_dns"] = "|".join([x.get("hostname",'') for x in dct["ti_data.passive_dns"]])
    dct["ti_data.passive_dns"] = dct["ti_data.passive_dns"].replace("&","&1").replace(",","&2")
    
    dct["ti_data.malware"]     = "|".join(x.get("malware",'') for x in dct["ti_data.malware"])
    dct["ti_data.malware"]     = dct["ti_data.malware"].replace("&","&1").replace(",","&2") 
    
    dct["ti_data.blacklist"]   = "|".join(dct.get("ti_data.blacklist",''))
    dct["ti_data.blacklist"]   = dct["ti_data.blacklist"].replace("&","&1").replace(",","&2")
    
    dct["ti_data.url_list"]    = "|".join([x.get("url",'') for x in dct["ti_data.url_list"]])
    dct["ti_data.url_list"]    = dct["ti_data.url_list"].replace("&","&1").replace(",","&2")
    
    dct["ti_data.dga"]                 = "|".join([x.get('dga','') for x in dct["ti_data.dga"]])
    dct["ti_data.dga"]                 = dct["ti_data.dga"].replace("&","&1").replace(",","&2") 
    
    dct["ti_data.ransomwaretracker"]   = "|".join(["RW_IP_Blocklist"+"^"+x.get("RW_IP_Blocklist",'')+"+"+"domain"+"^"+x.get("domain",'')+"+"+"desc"+"^"+x.get("desc",'') for x in dct["ti_data.ransomwaretracker"]])
    dct["ti_data.ransomwaretracker"]   = dct["ti_data.ransomwaretracker"].replace("&","&1").replace(",","&2")
    
    dct["fp_remarks"]                  = "|".join(dct.get("fp_remarks",''))
    dct["fp_remarks"]                  = dct["fp_remarks"].replace("&","&1").replace(",","&2")
    
    
    dct["anomalies"]                   = "|".join([x.get("msg",'')+"^"+x.get("note",'') for x in dct["anomalies"]])
    dct["anomalies"]                   = dct["anomalies"].replace("&","&1").replace(",","&2")
    
    dct["ti_data.malware_hashes"]      = "|".join([x.get("hash",'')+"^"+x.get("timestamp",'') for x in dct.get("ti_data.malware_hashes",'')])
    
    dct["ti_data.saniturl"] = "|".join([x.get("malware","")+"^"+x.get("filetype","")+"^"+x.get("url","") for x in dct["ti_data.saniturl"]])
    dct["ti_data.saniturl"] = dct["ti_data.saniturl"].replace("&","&1").replace(",","&2")
    
    dct["aso"]              = dct["aso"].replace("&","&1").replace(",","&2")
    dct["location"]         = dct["location"].replace("&","&1").replace(",","&2")
    
    if dct.get("OTX.url_list", ''):
        dct["OTX.url_list"] = "|".join([x["url"] for x in dct["OTX.url_list"]])
        dct["OTX.url_list"] = dct["OTX.url_list"].replace("&","&1").replace(",","&2")
    else:
        dct["OTX.url_list"] = ""
        
    if not dct.get("OTX.whois", ''):
        dct["OTX.whois"] = ""
    else:
        dct["OTX.whois"] = dct["OTX.whois"].replace("&","&1").replace(",","&2")
        
    if not dct.get("OTX_enabled", ''):
        dct["OTX_enabled"] = ""
        
    del dct["ti_data.tags"]
    del dct["ti_data.geo_info_mmdb"]
    return dct
```
